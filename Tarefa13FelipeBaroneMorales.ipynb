```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

br = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/tabelaBrasileirao2018.csv')

br
```

```python
plt.figure()
plt.title('Público do Corinthians quando mandante')
br.query("Mandante == 'Corinthians'")['Público'].hist()
plt.xlabel('Público Corinthians')
plt.show()

print('A média do público Corinthiano é: %.2f' % br.query("Mandante == 'Corinthians'")['Público'].median())
```
