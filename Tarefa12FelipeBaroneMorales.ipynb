Exercicio 1

```python
import numpy as np
import matplotlib.pyplot as plt

v = 10
y = 0
t = 0

for i in range(0, 250):
  v = v * (i - 1) + 0.01 * (-9.81)


for i in range(0, 250):
  y = y * (i - 1) + 0.01 * v * (i - 1)


for i in range(0, 250):
  t = t * (i - 1) + 0.01

plt.ylabel("y")
plt.xlabel("t")
plt.plot(y , t, color='blue')
plt.show()
plt.ylabel("v")
plt.xlabel("t")
plt.plot(v, t, color = 'red')



plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 10, 10000) # 100000 n√∫meros, de 0 a 10

for i in range(0, 10000):
  x = (4/(np.pi)) * (np.sin(((2*i-1)*t))/(2*i-1))
  if i == 10:
    plt.ylabel("x")

    plt.xlabel("t")

    plt.plot(t, x, color='green')

  if i == 100:
    plt.ylabel("x")

    plt.xlabel("t")

    plt.plot(t, x, color='blue')

  if i == 1000:
    plt.ylabel("x")

    plt.xlabel("t")

    plt.plot(t, x, color='red')

  if i == 10000:
    plt.ylabel("x")

    plt.xlabel("t")

    plt.plot(t, x, color='yellow')

plt.show()

```
