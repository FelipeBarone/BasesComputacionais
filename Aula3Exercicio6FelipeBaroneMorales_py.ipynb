Exercício 1

```python
import matplotlib.pyplot as plt
import numpy as np

e = np.e
pi = np.pi

t = np.arange(0, 10, 0.05)

v = (e**(-0.5*t)) * np.cos(2*pi*0.8*t)

plt.figure()
plt.plot(t, v)
plt.show
```

Exercício 2

```python
C = np.linspace(-20, -100, 100)
F = (9/5) * C + 32

plt.xlabel("C")

plt.ylabel("F")

plt.plot(C, F, color= 'red')
plt.show
```

Exercício 3

```python
a = np.arange(1994, 2022, 1)
hb = 800000 * np.e ** ((a - 1994)/40)

plt.xlabel("Ano")

plt.ylabel("Habitantes")

plt.plot(a, hb, color= 'red')
plt.show
```
